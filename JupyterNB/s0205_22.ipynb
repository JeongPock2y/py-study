```python
 df = pandas.read_csv('auto-mpg.csv')
cyl_size = (df['cylinders'] - df['cylinders'].min()) * 20 + 20

pyplot.style.use('ggplot')
df.plot(kind = 'scatter', x = 'weight', y = 'mpg', marker = '+', figsize = (7, 3),
        cmap = 'viridis', c = cyl_size, s = 50, alpha = 0.3)
pyplot.title('Auto-MPG weight-mpg Scatter Plot')
pyplot.show()
```

```python
df_origin = df['origin'].value_counts() / df['origin'].count() * 100
df_origin.rename({1:'USA', 2:'EU', 3:'JPN'}, inplace = True)
df_origin.plot(kind = 'pie', figsize = (5, 4), autopct = '%.1f%%',
               startangle = 10, colors = ['chocolate', 'bisque', 'cadetblue'])
pyplot.title('자동차 생산지역', size = 12)
pyplot.legend(loc = 'best', fontsize = 8)
pyplot.show()
```

```python
import seaborn
import pandas

titanic = seaborn.load_dataset('titanic')
titanic
```

```python
omit = abs(titanic.count() - len(titanic.index))
print("누락된 컬럼들과 수량:")
print(omit[omit > 0])
```

```python
# isnull()  : 비었는가? NaN,None -> 1, 유효값 -> 0
# notnull() : 비어있지 않은가? 유효값 -> 1, NaN,None -> 0
omit = titanic.isnull().sum()
print("누락된 컬럼들과 수량:")
print(omit[omit > 0], end = '\n\n')
print(titanic.count())
```
