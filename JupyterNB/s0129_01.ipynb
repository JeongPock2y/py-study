```python
import pandas

dict_data = {'수학':[90, 80,  70, 60], '영어':[ 98, 89, 95, 90],
             '음악':[85, 95, 100, 90], '체육':[100, 90, 90, 80]}    
df = pandas.DataFrame(dict_data, index = ['서준', '우현', '인아', '지훈'])
df
```

```python
df2 = df.copy()
df2.drop(['수학', '음악'], axis = 'columns', inplace = True)
df2.drop('우현', axis = 'index', inplace = True)
df2
```

```python
# DataFrame.drop('이름')    기본적으로 행(axis = 0, 'index')의 이름을 제거한다.
# 추가 옵션으로 axis = 1 / 'columns' 을 지정하여 열(columns)의 이름을 제거할 수 있다.
df2 = df.copy()
df2.drop('인아', inplace = True)
df2.drop('영어', axis = 1, inplace = True)
df2
```

```python
# DataFrame['column명'] = 값   : column명이 있을경우 기존 값이 변경,
#                                column명이 없을경우 새로운 컬럼이 생성되고, 값입력
# = 값                   : 모든 index에 동일한 값으로 입력
# = [값1, 값2, 값3, ...] : 모든 index에 별도의 값이 입력 (단, 갯수가 일치해야한다.)
df2 = df.copy()
df2['음악'] = 90
df2['역사'] = [51, 52, 53, 54]
df2
```

```python
# DataFrame.loc['idx명'] = 값   : idx명이 있을경우 기존 값이 변경,
#                                 idx명이 없을경우 새로운 행이 생성되고, 값입력
df2 = df.copy()
df2.loc['우현'] = 70
df2.loc['지희'] = [91, 92, 93, 94]
df2
```

```python
# DataFrame.loc[idx명, column명]        <- 원소를 조회
# DataFrame.loc[idx명, column명] = 값   <- 원소를 변경(수정)
df2 = df.copy()
print(df2.loc['우현', '영어'])
df2.loc['우현', '영어'] = 100
print(df2.loc['우현', '영어'])
df2.loc['지훈', ['영어', '체육']] = [91, 92]
df2
```

```python
list_data = [[100, 120, 150, 180],
             [ 80,  85,  90,  90],
             [100,  95,  90,  80],
             [130, 180, 200, 250]]
df = pandas.DataFrame(list_data, columns = [1991, 1992, 1993, 1994],
                      index = ['InCheon', 'Busan', 'KwangJu', 'Seoul'])
# DataFrame.transpose() : DataFrame객체의 행/열 전환된 객체를 반환
# DataFrame.T           :  DataFrame객체 원본의 행/열 전환된 결과를 조회
df.T.plot()
```

```python
dict_data = {'이름':['서준', '우현', '인아', '진수'],
             '수학':[90, 80,  70, 60], '영어':[ 98, 89, 95, 90],
             '음악':[85, 95, 100, 90], '체육':[100, 90, 90, 80],
             '국어':[80, 80,  80, 80]}
df = pandas.DataFrame(dict_data)
# DataFrame.set_index('column명') : 특정한 columns을 index로 지정
# DataFrame.reset_index()         : index 축을 columns으로 복구
# df.set_index('이름', inplace = True)  ,  df = df.set_index('이름')
df = df.set_index(['이름'])
df
```

```python
# NaN(Not a Number) : 조사되지 않은 데이터(결측치)
# reindex([새로운 index의 배열]) : 기존 index순서와 상관이 새로운 순서로 배열하고,
#                                  없었던 index가 있을경우, index가 추가되고,
#                                  값은 NaN으로 입력, fill_value = 새로운값 변경가능
df2 = df.reindex(['인아', '우현', '수민', '진수', '서준', '예진'], fill_value = 50)
df2
```

```python
df3 = df2.copy()
df3.reset_index()
```

```python
# 정렬 : index명으로 오름차순/내림차순 , 특정한 column명의 값들을 기준으로 정렬
# DataFrame.sort_index()                   : index명을 기준으로 정렬
# DataFrame.sort_values(by = ['column명']) : column명의 값을 기준으로 정렬
# 정렬의 기본은 오름차순(ascending = True), 내림차순(ascending = False)
df3 = df2.copy()
#df3.sort_index(ascending = False)
df3.sort_values(by = ['체육', '수학'], ascending = False)
```
