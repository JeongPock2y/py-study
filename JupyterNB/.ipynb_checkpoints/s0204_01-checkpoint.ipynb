```python
import pandas as pd

def grade(avg):
    if avg >= 90:   grd = 'A'
    elif avg >= 80: grd = 'B'
    elif avg >= 70: grd = 'C'
    elif avg >= 60: grd = 'D'
    else:           grd = 'F'
    return grd
```

```python
import pandas as pd
for i in range(1,4):
    #file1 = 'C:\doit\py-study\JupyterNB\\class_{i}.csv'
    df = pd .read_csv('C:\doit\py-study\JupyterNB\\class_{i}.csv', encoding= 'euckr',header=None,
    names= ['name','addr','kor','eng','math','sci','his'],index_col = 'name')
    subs=['kor','eng','math','sci','his']
    df['avg']=df[subs].sum(axis=1) / len(subs)
    df['grd']=df['avg'].map(grade)
    df.to_excel('C:\doit\py-study\JupyterNB\\class_{i}_avg_grd.xlsx')
 
```
