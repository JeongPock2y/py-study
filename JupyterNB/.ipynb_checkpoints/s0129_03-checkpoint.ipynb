```python
def grade(avg):
    if avg >= 90:   grd = 'A'
    elif avg >= 80: grd = 'B'
    elif avg >= 70: grd = 'C'
    elif avg >= 60: grd = 'D'
    else:           grd = 'F'
    return grd

import pandas
# sheet_name = '시트명' : 특정 시트를 불러올 수 있다. 기본값은 첫번째 시트이다.
df = pandas.read_excel('C:\doit\py-study\JupyterNB\students.xlsx', sheet_name = '1반', index_col = '이름')
df['평균'] = df.loc[:,'국어':'역사'].sum(axis = 'columns') / 5
df
```
